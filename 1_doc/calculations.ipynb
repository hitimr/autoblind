```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mass = 2.2  # kg
F = mass*9.81 

def torque_required_Ncm(d_vals, gear_ration=1, efficiency=1):
    return [(d/2)*F/gear_ration/efficiency for d in d_vals]


d_vals_mm = np.arange(start=5, stop=50, step=1)
d_vals_cm = d_vals_mm/10
torque_vals_Ncm =torque_required_Ncm(d_vals_mm, gear_ration=20, efficiency=0.5)

plt.plot(d_vals_mm, torque_vals_Ncm)
plt.xlabel("Wellendurchmesser [mm]")
plt.ylabel("Drehmoment [Ncm]")
plt.grid()
```
